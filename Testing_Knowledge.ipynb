```python
#Task 2
def largest_of_three(a:float, b:float, c:float):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    elif c >= b and c >= a:
        return c

largest_of_three(3.2, 1.2, 11.2)
```

```python
#Task 3
def grade_student(score:float):
    try:

        if score >= 90:
            return 'A'
        elif 80 <= score <= 89:
            return 'B'
        elif 70 <= score <= 79:
            return 'C'
        elif 60 <= score <= 69:
            return 'D'
        elif score <= 59:
            return 'F'
    except TypeError:
        return 'Not a Valid Data Type for score'


grade_student(12)
```

```python
#Task 5
def classify_temperature(temp):
    try:

        if temp < 0:
            return "Freezing"
        elif temp >= 0 and temp <= 15:
            return "Cold"
        elif temp >= 16 and temp <= 25:
            return "Mild"
        else:
            return "Hot"
    except TypeError:
        return "Not Valid input"

classify_temperature(5)
```

```python
#Task 1
def sum_to_n(n:int):
    if n <= 0:
        return 'Enter a Positive Integer for n'
    storing_list = []
    for i in range(n+1):
        storing_list.append(i)
    x = sum(storing_list)
    return x



sum_to_n(0)
```

```python
def count_down(start):
    if start != int(start) or start < 0:
        print('Provide a valid positive integer for start')
    else:
        try:
            while start > 0:
                print(start)
                start = start - 1
            else:
                print('Blast off!')
        except TypeError:
            return 'Provide Valid input for start'
```

```python
def reverse_list(items:list):                 #defines function name also I made sure to allow input for list only
    if items == []:                           #handled edge case for when there is nothing in the list
        print('Provide a complete list')      #if there is only 1 item in the list we simply just return it since same
    elif len(items) == 1:
        return items
    else:                                     #here I start with an empty list to append items from orignial list
        new_reversed_list = []
        for i in items:                           #here we go through each element in items and I say I want to append an item
            new_reversed_list.append(items.pop()) #into the new list that will be reversed and I simply say take the last item
        new_reversed_list.append(items[0])        #since last item removed from original list we iterate again find last item again
        return new_reversed_list                  #then the next line returns the list but the final iterated list still has one item
                                                  #then we append that last item into the reversed list and return all of it


reverse_list(['apple'])
```

```python
#Task 5:

def print_multiplication_table(n:int):
    try:
        if n != int(n) and n > 1:
            print('Enter Valid integer for n')
        elif n < 1:
            print('Enter Valid integer for n')
        else:
            for i in range(1,n+1):
                for j in range(1,n+1):
                    print(i * j , end=' ')
                print()
    except ValueError:
        pass          #don't need to handle this but just wanted to add a try block

```

```python
print_multiplication_table(3)
```

```python
def fact_n(n: int):
    try:
        prod = 1
        if n == 0:
            return 1
        elif n < 0:
            return f'{n} not a positive integer'
        else:
            for i in range(1,n+1):
                prod = prod * i
            return prod
    except TypeError:
        return f'{n} not valid integer format'

fact_n(3)
```
